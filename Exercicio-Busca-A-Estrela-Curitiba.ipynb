# Exercicio

Vamos construir um Grafo para saber qual é a melhor distância entre a cidade de PORTO UNIÃO para CURITIBA, utilizando o método de Busca A Estrela A*.

```python
# Primeiro, vamos criar a classe chamada "Vertice", que irá representar as cidades mostradas no mapa

class Vertice:
    # Construtor
    def __init__(self, rotulo, distancia_objetivo):
        # Nome da cidade
        self.rotulo = rotulo
        # Mostrar se a cidade já foi visitada
        self.visitado = False
        # A distância até a cidade objetivo (linha reta)
        self.distancia_objetivo = distancia_objetivo
        # Lista que irá mostrar as cidades adjacentes
        self.adjacentes = []
    
    # Função que irá adicionar na lista "adjacentes" as cidades vizinhas
    def adiciona_adjacente(self, adjacente):
        self.adjacentes.append(adjacente)
    
    # Função que irá mostrar na tela os itens dentro da lista "adjacentes"
    def mostra_adjacentes(self):
        for i in self.adjacentes:
            print(i.vertice.rotulo, i.custo)
```

```python
# Agora, vamos criar uma classe chamada "Adjacente", onde irá representar a ligação entre as cidades

class Adjacente:
    def __init__(self, vertice, custo):
        # Nome da cidade
        self.vertice = vertice
        # Custo de distância
        self.custo = custo
        
        # Atributo que irá calcular a soma da distância (linha reta) e o custo para chegar na cidade
        self.distancia_aestrela = vertice.distancia_objetivo + self.custo
```

# Vetor Ordenado (adjacente)

```python
import numpy as np

# Vamos criar uma classe chamada "VetorOrdenado", onde ficará estabelecido o limite das cidades adjacentes
class VetorOrdenado:
    
    def __init__(self, capacidade):
        self.capacidade = capacidade
        self.ultima_posicao = -1
        
        # Mudando no tipo de dados
        self.valores = np.empty(self.capacidade, dtype=object)
    
    # Referência para o vértice e comparação com a distância A*
    def insere(self, adjacente):
        if self.ultima_posicao == self.capacidade - 1:
            print('Capacidade máxima atingida!')
            return
        
        posicao = 0
        
        for i in range(self.ultima_posicao + 1):
            
            posicao = i
            
            if self.valores[i].distancia_aestrela > adjacente.distancia_aestrela:
                break
            
            if i == self.ultima_posicao:
                posicao = i + 1
        
        x = self.ultima_posicao
        
        while x >= posicao:
            self.valores[x + 1] = self.valores[x]
            
            x -= 1
            
        self.valores[posicao] = adjacente
        self.ultima_posicao += 1
    
    
    def imprime(self):
        if self.ultima_posicao == -1:
            print('O vetor está vazio!')
        else:
            for i in range(self.ultima_posicao + 1):
                print(i, ' - ', self.valores[i].vertice.rotulo, ' - ',
                     self.valores[i].custo, ' - ',
                     self.valores[i].vertice.distancia_objetivo, ' - ',
                     self.valores[i].distancia_aestrela)
```

# Fazendo a Busca AEstrela

```python
class AEstrela:
    def __init__(self, objetivo):
        self.objetivo = objetivo
        self.encontrado = False
    
    
    def buscar(self, atual):        
        print('--------------')
        print('Atual: {}'.format(atual.rotulo))
        atual.visitado = True
        
        if atual == self.objetivo:
            self.encontrado = True
        else:
            vetor_ordenado = VetorOrdenado(len(atual.adjacentes))
            
            for adjacente in atual.adjacentes:
                if adjacente.vertice.visitado == False:
                    adjacente.vertice.visitado = True
                    vetor_ordenado.insere(adjacente)
            
            vetor_ordenado.imprime()
            
            if vetor_ordenado.valores[0] != None:
                self.buscar(vetor_ordenado.valores[0].vertice)
```

```python
# Definindo a cidade objetivo (Curitiba)
busca_aestrela = AEstrela(grafo.curitiba)

# Definindo a cidade original (Porto União)
busca_aestrela.buscar(grafo.porto)
```
